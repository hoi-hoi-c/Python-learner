```python
def userinput():
    a=input('평문 입력: ')
    b=int(input('비밀키 입력: '))
    c=input('암호화/복호화?: ')
    return a,b,c

def add_key():
    for i in range(len(pwd)):
        newpwd_num=ord(pwd[i])+keynum
        if newpwd_num>122:
            newpwd_num=newpwd_num-26
        newpwd_list.append(chr(newpwd_num))
    return newpwd_list

def sub_key():
    for i in range(len(pwd)):
        newpwd_num=ord(pwd[i])-keynum
        if newpwd_num<97:
            newpwd_num=newpwd_num+26
        newpwd_list.append(chr(newpwd_num))
    return newpwd_list

def encoding(r):
    r1=''
    for i in range(len(r)):
        r1+=r[i]
    return r1

#메인코드
newpwd_list=[]

pwd,keynum,choice=userinput()
#pwd=pwd.lower()
if choice=='암호화':
    newpwd=add_key()
elif choice=='복호화':
    newpwd=sub_key()
newpwd=encoding(newpwd)
print('암호문:%s'%newpwd)
```
