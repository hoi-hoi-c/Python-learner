```python
#전역 변수, 디폴트 인자
#전역변수 area에 계산값을 저장하려면?
def calculate_area(radius):
    global area
    area=3.14*radius**2
    return

area=0
r=float(input("원의 반지름: "))
calculate_area(r)
print(area)
```

```python
def para_func(v1,v2,v3=0):
    result=0
    result=v1+v2+v3
    return result

hap=0

#메인 코드 부분
hap=para_func(10,20)
print('매개변수가 2개인 함수를 호출한 결과 ==>%d'%hap)
hap=para_func(10,20,30) #매개변수에 지정하지 않으면 default값으로 함수 돌림, 매개변수 지정되면 지정된 내용으로 바뀜.
print('매개변수가 3개인 함수를 호출한 결과 ==>%d'%hap)
```

```python
#매개변수의 개수를 지정하지 않고 사용하려면
def para_func(*para):
    result=0
    for num in para:
        result=result+num
    return result

hap=0

hap=para_func(10,20)
print('매개변수가 2개인 함수를 호출한 결과 ==>%d'%hap)
hap=para_func(10,20,30)
print('매개변수가 3개인 함수를 호출한 결과 ==>%d'%hap)
hap=para_func(10,20,30,40,50)
print('매개변수가 5개인 함수를 호출한 결과 ==>%d'%hap)
```

```python
#*args는 다른 매개변수와 함께 사용할 수도 있다.
def add_mul(choice,*args):
    if choice=='add':
        result=0
        for i in args:
            result=result+i
    elif choice=='mul':
        result=1
        for i in args:
            result=result*i
    return result

result=add_mul('add',1,2,3,4,5)
print(result)
result=add_mul('mul',1,2,3,4)
print(result)
```

```python
#계산기
def userwrite():
    a=int(input('첫 번째 정수를 입력하세요: '))
    b=int(input('두 번째 정수를 입력하세요: '))
    cal=input('기호를 입력하세요(+,-,*,/,all,who,end): ')
    return a,b,cal

def calculator(cal):
    if cal=='+': print(a+b)
    elif cal=='-': print(a-b)
    elif cal=='*': print(a*b)
    else:
        if b!=0: print(a/b)
        else: print('분모가 0입니다.')

def allfunc():
    print('addition result:',a+b)
    print('substraction result:',a-b)
    print('multiplication result:',a*b)
    if b!=0: print('division result:',a/b)
    else: print('분모가 0입니다.')
def whofunc():
    print('-'*20)
    print('경영대학, 2014026826, 채정회')
    print('-'*20)

while True:
    a,b,cal=userwrite()
    if cal in ['+','-','*','/']:
        calculator(cal)
    elif cal=='all':
        allfunc()
    elif cal=='who':
        whofunc()
    else:
        break
```
