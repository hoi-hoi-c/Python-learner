```python
#Lambda 함수
def  square2(x):
    return x**2
print(square2(5))

square=lambda x:x**2
print(square(5))
#-----
def add(x,y):
    return x+y
print(add(10,20))

add2=lambda x,y:x+y
print(add2(10,20))
```

```python
def hap(num1,num2):
    res=num1+num2
    return res
print(hap(10,20))

hap2=lambda num1,num2:num1+num2
print(hap2(10,20))
```

```python
hap3=lambda num1=10,num2=20:num1+num2
print(hap3())
print(hap3(100,200))

print(lambda num3:num3*2,10)
```

```python
#리스트에 모두 10을 더하는 코드
myList=[1,2,3,4,5]
def add10(num):
    return num+10

for i in range(len(myList)):
    myList[i]=add10(myList[i])
print(myList)
```

```python
#map(): 각 원소를 주어진 수식에 따라 변형하여 새로운 리스트를 반환
myList=[1,2,3,4,5]
add10=lambda num:num+10
myList=list(map(add10,myList))
print(myList)

myList=[1,2,3,4,5]
myList=list(map(lambda num:num+10,myList))
print(myList)
```
