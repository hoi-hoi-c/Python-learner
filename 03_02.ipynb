```python
def star():
    i=1
    while i<=7:
        print('*'*i)
        i+=1

star()
star()
```

```python
def star(eng):
    i=1
    while i<=7:
        print(eng*i)
        i+=1

star('K')
star('A')
star('B')
```

```python
def star(eng):
    i=6
    while i>=1:
        print(eng*i)
        i-=1

star('O')
star('T')
```

```python
def starup(eng):
    i=1
    while i<=7:
        print(eng*i)
        i+=1

def stardown(eng):
    i=6
    while i>=1:
        print(eng*i)
        i-=1
        
starup('X-mas')
stardown('merry')
```
